```python
#Data processing 

# Read the input CSV file into a Pandas DataFrame
input_df = pd.read_csv("./users_example.csv")

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=["user", "xvs_balance"])

# Iterate over the "users" column and perform some operation on each value
for user in input_df["users"]:
    
    # Replace this with whatever operation you want to perform on each user
    result = vault_contract.functions.getUserInfo(xvs, 0 , Web3.toChecksumAddress(user)).call()[1] / 1e18
    
    # Add the result to the result DataFrame
    result_df = result_df.append({"user": user, "xvs_balance": result}, ignore_index=True)

# Save the result DataFrame to a new CSV file
result_df.to_csv("user_xvs_balance.csv", index=False)

```
